```python
import pandas as pd
import yfinance as yf
```

```python
bolsa = '^BVSP'
inicio = '2003-12-01'
dolar = 'BRL=X'
```

```python
bovespa = yf.download(bolsa, inicio, multi_level_index=False)["Close"]
```

```python
bolsa = pd. DataFrame(bovespa)
bolsa
```

```python
USD = yf.download(dolar, inicio, multi_level_index=False)['Close']
```

```python
usd = pd.DataFrame(USD)
usd
```

```python
bv_em_dolar = pd.merge(bolsa, usd, how='inner')
```

```python
bv_em_dolar
```

```python
df = yf.download(tickers=['^BVSP', 'BRL=X'], start=inicio, multi_level_index=False)
```

```python
df_1 = df.Close
```

```python
df_2 = df_1.dropna()
```

```python
df_2['BV_em_dolar']= df_2['^BVSP']/df_2['BRL=X']
```

```python
df_3 = df_2.iloc[0]/df_2
```

```python
df_3.plot()
```

```python
df_2_renamed = df_2.rename(columns={'BRL=X': 'Dolar', '^BVSP': 'Bovespa', 'BV_em_dolar': 'Bovespa Dolar'})
```

```python
df_2_renamed[['Bovespa', 'Bovespa Dolar']].plot()
```
